```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
file_path = "/Restaurant reviews.csv"
reviews_df = pd.read_csv(file_path)
```

```python
reviews_df.head()
```

```python
reviews_df.info
```

```python
reviews_df['Review'].fillna('', inplace=True)
```

```python
def preprocess_text(text):
    if isinstance(text, str):
        # Tokenize text
        tokens = word_tokenize(text)
        # Convert to lower case
        tokens = [word.lower() for word in tokens]
        # Remove punctuation
        tokens = [word for word in tokens if word.isalpha()]
        # Remove stop words
        stop_words = set(stopwords.words('english'))
        tokens = [word for word in tokens if not word in stop_words]
        # Lemmatize
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]
        return ' '.join(tokens)
    else:
        return ''
```

```python
reviews_df['Cleaned_Review'] = reviews_df['Review'].apply(preprocess_text)
```

```python
reviews_df['Cleaned_Review'].head()
```

```python
def is_numeric(value):
    try:
        float(value)
        return True
    except ValueError:
        return False
```

```python
reviews_df = reviews_df[reviews_df['Rating'].apply(is_numeric)]
```

```python
reviews_df['Rating'] = reviews_df['Rating'].astype(float)
```

```python
reviews_df = reviews_df.dropna(subset=['Rating'])
```

```python
reviews_df['Rating'] = reviews_df['Rating'].round().astype(int)
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(reviews_df['Cleaned_Review'])
```

```python
X
```

```python
reviews_df['Sentiment'] = reviews_df['Rating'].apply(lambda x: 1 if x >= 4 else (0 if x <= 2 else -1))
```

```python
reviews_df = reviews_df[reviews_df['Sentiment'] != -1]

X = vectorizer.fit_transform(reviews_df['Cleaned_Review'])
y = reviews_df['Sentiment']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

model = MultinomialNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
X_train.shape,X_test.shape, y_train.shape,y_test.shape
```

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
def predict_sentiment(text):
    text_processed = preprocess_text(text)
    text_vectorized = vectorizer.transform([text_processed])
    prediction = model.predict(text_vectorized)
    return "Positive" if prediction[0] == 1 else "Negative"
```

```python
sample_review = "The food was absolutely wonderful, from preparation to presentation, very pleasing."
print(f"Review: {sample_review}")
print(f"Sentiment: {predict_sentiment(sample_review)}")
```
